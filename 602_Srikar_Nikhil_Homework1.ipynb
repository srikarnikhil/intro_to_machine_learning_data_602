# Generate a normally distributed data, with mean=3 and standard deviation=2, for 10,000 observations

```python
mu, sigma = 3, 2 
```

```python
no_observations = 10000
```

```python
random_data = np.random.normal(mu, sigma, no_observations)
```

# Plot the normal distribution overlaid on top of a histogram in matplotlib 

```python
# plotting the population data as histogram
hist_values, bins, bar_container = plt.hist(random_data,bins=30,density=True)

# getting the min and max values of the plot
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)

# plotting the overlaid
plt.plot(x, p, 'k', linewidth=2,color = 'r')
plt.show()
```

# Create a sub-sample of this population with mean of  1.5, a sigma of 2.0, and 200 observations

```python
sample_mean, sample_sigma = 1.5, 2
```

```python
no_sample_observations = 200
```

```python
# generating the sample data
sample = np.random.normal(sample_mean, sample_sigma, no_sample_observations)
```

```python
# plotting the population data as histogram
hist_values, pop_bins, bar_container = plt.hist(random_data,bins=30,density=True,alpha=0.1,label='population')
xticks(range(-4,11,2))

# getting the min and max values of the plot
xmin_pop, xmax_pop = plt.xlim()
x_pop = np.linspace(xmin_pop, xmax_pop, 100)
p_pop = norm.pdf(x_pop, mu, sigma)

# plotting the overlaid
plt.plot(x_pop, p_pop, 'k', linewidth=2,color = 'b',label='population')

# plotting the sample data as histogram
count, sample_bins, bar_container_ignored = plt.hist(sample, bins = 30, alpha=0.1, density=True,label='sample')

# getting the min and max values of the plot
xmin, xmax = plt.xlim()
x_sam = np.linspace(xmin, xmax, 100)
p_sam = norm.pdf(x_sam, sample_mean, sample_sigma)

# plotting the overlaid
plt.plot(x_sam, p_sam, 'k', linewidth=2,color = 'r',label='sample')
plt.legend()
plt.show()
```

# adding confidence interval lines

```python
# plotting the population data as histogram
hist_values, pop_bins, bar_container = plt.hist(random_data,bins=30,density=True,alpha=0.1,label='population')
xticks(range(-4,11,2))
xmin_pop, xmax_pop = plt.xlim()
x_pop = np.linspace(xmin_pop, xmax_pop, 100)
p_pop = norm.pdf(x_pop, mu, sigma)

# plotting the overlaid
plt.plot(x_pop, p_pop, 'k', linewidth=2,color = 'b',label='population')


# plotting the sample data as histogram
count, sample_bins, bar_container_ignored = plt.hist(sample, bins = 30, alpha=0.1, density=True,label='sample')

# getting the min and max values of the plot
xmin, xmax = plt.xlim()
x_sam = np.linspace(xmin, xmax, 100)
p_sam = norm.pdf(x_sam, sample_mean, sample_sigma)

# plotting the overlaid
plt.plot(x_sam, p_sam, 'k', linewidth=2,color = 'r',label='sample')

#calculating the confidence intervals
ci = norm.interval(alpha=0.95,
                 loc=sample_mean,
                 scale=sample_sigma)

# Adding  vertical lines at the confidence intervals with color green
plt.axvline(ci[0],color='g')
plt.axvline(ci[1],color='g')
plt.legend()
plt.show()



```

# Create a graph that shows the null and hypotheses, the type 1 and 2 errors

```python
# plotting the population data as histogram
hist_values, pop_bins, bar_container = plt.hist(random_data,bins=30,density=True,alpha=0.1,label='population')
xticks(range(-4,11,2))

# getting the min and max values of the plot
xmin_pop, xmax_pop = plt.xlim()
x_pop = np.linspace(xmin_pop, xmax_pop, 100)
p_pop = norm.pdf(x_pop, mu, sigma)

# plotting the overlaid
plt.plot(x_pop, p_pop, 'k', linewidth=2,color = 'b',label='population')


# plotting the sample data as histogram
count, sample_bins, bar_container_ignored = plt.hist(sample, bins = 30, alpha=0.1, density=True,label='sample')

# getting the min and max values of the plot
xmin, xmax = plt.xlim()
x_sam = np.linspace(xmin, xmax, 100)
p_sam = norm.pdf(x_sam, sample_mean, sample_sigma)

# plotting the overlaid
plt.plot(x_sam, p_sam, 'k', linewidth=2,color = 'r',label='sample')

#calculating the confidence intervals
ci = norm.interval(alpha=0.95,
                 loc=sample_mean,
                 scale=sample_sigma)

# Adding  vertical lines at the confidence intervals with color green
plt.axvline(ci[0],color='g')
plt.axvline(ci[1],color='g')

#Filling the area between two horizontal curves
range_1 = np.arange(-4,ci[0],0.01) # from -4 to left confidence interval
plt.fill_between(x=range_1, 
                 y1= norm.pdf(range_1,loc=sample_mean,scale=sample_sigma) ,
                 facecolor='red',
                 alpha=0.35)

range_2 = np.arange(ci[1],7.5,0.01) # from right confidence interval to 7.5
plt.fill_between(x=range_2, 
                 y1= norm.pdf(range_2,loc=sample_mean,scale=sample_sigma) ,
                 facecolor='red',
                 alpha=0.5)

range_3 = np.arange(ci[0],ci[1],0.01) # in between confidence intervals
plt.fill_between(x=range_3, 
                 y1= norm.pdf(range_3,loc=mu, scale=sigma) ,
                 facecolor='blue',
                 alpha=0.5)
plt.legend()
# inserting the texts
plt.text(x=-4.2, y=0.01, s= "Type 1 Error")
plt.text(x=1.7, y=0.015, s= "Type 2 Error")
plt.text(x=0.1, y=0.18, s= "Null Hypothesis")
plt.text(x=6, y=0.05, s= "Alternative")
plt.text(x=6.25, y=0.01, s= "Type 1 Error")
```

# Compute the z-score

```python
z_score=(sample_mean-mu)/sigma 
z_score
```

```python
#calculating the p value(two tailed)
p = norm.sf(abs(z_score)*2)
```

# What is the p-value and do you reject it?

```python
print(p)
```

```python
p>0.05
```

```python
def check_rejection(p_value,alpha):
    if p_value<alpha: 
         print('Since p value: {} is less than alpha: {} (significance level), we reject the Null Hypothesis'.format(p_value,alpha)) 
    else: 
         print('Since p value: {} is greater than alpha:{} (significance level), we do not reject the Null Hypothesis'.format(p_value,alpha))
```

```python
check_rejection(p,0.05)
```
